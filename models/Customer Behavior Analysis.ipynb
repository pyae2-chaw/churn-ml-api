```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

sns.set(style="whitegrid")
```

```python
def normalize_columns(df):
    df.columns = [re.sub(r'[^a-z0-9]', '_', col.lower()) for col in df.columns]
    return df
```

```python
df = pd.read_csv('predicted_output.csv')
df = normalize_columns(df)
```

```python
# Gender distribution
if 'gender' in df.columns:
    sns.countplot(x='gender', data=df)
    plt.title("Gender Distribution")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.show()
```

```python
# Marital Status
if 'marital_status' in df.columns:
    sns.countplot(x='marital_status', data=df)
    plt.title("Marital Status")
    plt.xlabel("Status")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Number of Dependents
if 'number_of_dependents' in df.columns:
    sns.histplot(df['number_of_dependents'], bins=10, kde=False)
    plt.title("Number of Dependents")
    plt.xlabel("Dependents")
    plt.ylabel("Count")
    plt.show()
```

```python
# Income Distribution
if 'income' in df.columns:
    sns.histplot(df['income'], bins=30, kde=True)
    plt.title("Income Distribution")
    plt.xlabel("Income")
    plt.ylabel("Count")
    plt.show()
```

```python
# Education Level
if 'education_level' in df.columns:
    sns.countplot(x='education_level', data=df, order=df['education_level'].value_counts().index)
    plt.title("Education Level Distribution")
    plt.xlabel("Education Level")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Customer Segment
if 'customer_segment' in df.columns:
    sns.countplot(x='customer_segment', data=df)
    plt.title("Customer Segments")
    plt.xlabel("Segment")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Preferred Communication Channel
if 'preferred_communication_channel' in df.columns:
    sns.countplot(x='preferred_communication_channel', data=df)
    plt.title("Preferred Communication Channels")
    plt.xlabel("Channel")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```
