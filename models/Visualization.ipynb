```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your predicted dataset
df = pd.read_csv("predicted_output.csv")

# Optional: improve default style
sns.set(style="whitegrid")
```

```python
df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='churn_prediction', data=df)
plt.title("Churn Prediction Distribution")
plt.xlabel("Churn Prediction (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(x='gender', hue='churn_prediction', data=df)
plt.title("Churn by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title='Churn Prediction', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.histplot(df['churn_probability'], bins=30, kde=True)
plt.title("Distribution of Churn Probabilities")
plt.xlabel("Churn Probability")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x='churn_prediction', y='income', data=df)
plt.title("Income Distribution by Churn Prediction")
plt.xlabel("Churn Prediction")
plt.ylabel("Income")
plt.xticks([0, 1], ['No Churn', 'Churn'])
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix (Numeric Features)")
plt.show()
```
